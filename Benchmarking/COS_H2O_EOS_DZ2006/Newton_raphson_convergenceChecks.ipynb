```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import DiadFit as pf
from tqdm import tqdm
pf.__version__
```

```python
# Lets add some manually from the opposite of the inversion
P=np.linspace(1, 100, 40)
T_K=1400
molar_vol_0=pf.calculate_molar_volume_DZ2006(P_kbar=P, T_K=T_K, XH2O=0)
dens_0=pf.mol_vol_to_density(mol_vol=molar_vol_0, XH2O=0)
molar_vol_3=pf.calculate_molar_volume_DZ2006(P_kbar=P, T_K=T_K, XH2O=0.3)
dens_3=pf.mol_vol_to_density(mol_vol=molar_vol_3, XH2O=0.3)
molar_vol_5=pf.calculate_molar_volume_DZ2006(P_kbar=P, T_K=T_K, XH2O=0.5)
dens_5=pf.mol_vol_to_density(mol_vol=molar_vol_5, XH2O=0.5)
molar_vol_10=pf.calculate_molar_volume_DZ2006(P_kbar=P, T_K=T_K, XH2O=1)
dens_10=pf.mol_vol_to_density(mol_vol=molar_vol_10, XH2O=1)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (10,5))

ax1.plot(dens_0, P*1000, '.k', label='XH2O=0')
ax1.plot(dens_3, P*1000, '.r', label='XH2O=0.3')
ax1.plot(dens_5, P*1000, '.g', label='XH2O=0.5')
ax1.plot(dens_10, P*1000, '.c', label='XH2O=1')

ax2.plot(molar_vol_0, P*1000, '.k', label='XH2O=0')
ax2.plot(molar_vol_3, P*1000, '.r', label='XH2O=0.3')
ax2.plot(molar_vol_5, P*1000, '.g', label='XH2O=0.5')
ax2.plot(molar_vol_10, P*1000, '.c', label='XH2O=1')


# Now the algorithm to solve with Newton Raphson
dens=np.linspace(0.1, 2, 100)
mol_vol_P0=pf.density_to_mol_vol(density=dens, XH2O=0)
mol_vol_P03=pf.density_to_mol_vol(density=dens, XH2O=0.3)
mol_vol_P05=pf.density_to_mol_vol(density=dens, XH2O=0.5)
mol_vol_P10=pf.density_to_mol_vol(density=dens, XH2O=1)

P_0=pf.calculate_Pressure_DZ2006(density=dens, T_K=T_K, XH2O=0)
P_03=pf.calculate_Pressure_DZ2006(density=dens, T_K=T_K, XH2O=0.3)
P_05=pf.calculate_Pressure_DZ2006(density=dens, T_K=T_K, XH2O=0.5)
P_10=pf.calculate_Pressure_DZ2006(density=dens, T_K=T_K, XH2O=1)

ax1.plot(dens, P_0, '-k')
ax1.plot(dens, P_03, '-r')
ax1.plot(dens, P_05, '-g')
ax1.plot(dens, P_10, '-c')
# Now lets plot molar volume
ax2.plot(mol_vol_P0, P_0, '-k')
ax2.plot(mol_vol_P03, P_03, '-r')
ax2.plot(mol_vol_P05, P_05, '-g')
ax2.plot(mol_vol_P10, P_10, '-c')
ax1.set_xlabel('Bulk Density g/cm3')
ax1.set_ylabel('Pressure (bars)')
ax1.set_ylim([0, 100*1000])
ax2.set_ylim([0, 100*1000])
ax2.set_xlim([0, 70])
ax2.set_xlabel('Molar Vol')
ax2.set_ylabel('Pressure (bars)')
```
