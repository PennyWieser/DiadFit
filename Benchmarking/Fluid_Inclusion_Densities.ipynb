# Testing CoolProp vs. literature calculations of fluid inclusion pressures
- Here, we show that our estimates using CoolProp match literature etimates to convert CO2 density to depth

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import DiadFit as pf
```

## Test 1 - Yamamoto et al. 2007

## OK21 median pressure - Theirs is 7.06 kbar, upper is 8.11 kbar, lower is 7.1 kbar

```python
Ok21_Med=pf.convert_density_depth_Coolprop(T_K=1048+273.15, 
                                   Crust_dens_gcm3=2.7,
                CO2_dens_gcm3=1.005, output='df')
print('Median Pressure')
print(Ok21_Med['Pressure (kbar)'])

Ok21_Up=pf.convert_density_depth_Coolprop(T_K=1048+273.15, 
                                   Crust_dens_gcm3=2.7,
                CO2_dens_gcm3=1.023, output='df')
print('Up Pressure')
print(Ok21_Up['Pressure (kbar)'])

Ok21_Low=pf.convert_density_depth_Coolprop(T_K=1048+273.15, 
                                   Crust_dens_gcm3=2.7,
                CO2_dens_gcm3=0.986, output='df')
print('Low Pressure')
print(Ok21_Low['Pressure (kbar)'])
```
