```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import joblib
```

```python
data=pd.read_excel('Cell_fitting_params.xlsx', sheet_name='Sheet1')
```

```python
Pf_diad1_split_fwhm = np.poly1d(np.polyfit(data['Splitting'], 
                                data['Diad1_fwhm'],
                              6))
Px=np.linspace(102.5, 105, 100)
Py_diad1=Pf_diad1_split_fwhm(Px)


Pf_diad2_split_fwhm = np.poly1d(np.polyfit(data['Splitting'], 
                                data['Diad2_fwhm'],
                              6))
Px=np.linspace(102.5, 105, 100)
Py_diad2=Pf_diad2_split_fwhm(Px)
```

```python
plt.plot(data['Splitting'], data['Diad1_fwhm'], 'ok')
plt.plot(Px, Py_diad1, '-r')
plt.plot(data['Splitting'], data['Diad2_fwhm'], 'ob')

plt.plot(Px, Py_diad2, '-b')
```
