```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import DiadFit as pf
pf.__version__
```

```python
df_combo_voigt=pd.read_excel('Voigt.xlsx')
df_combo_pvoigt=pd.read_excel('PseudoVoigt.xlsx')
De=pd.read_excel('Fityk_results.xlsx', sheet_name='NEON')
```

```python
df_combo_pvoigt.columns
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,4))

ax1.plot(df_combo_voigt['deltaNe'],  De['Ne-m (cm-1)'], '.k')
ax1.plot([330.26, 330.3], [330.26, 330.3], '-r')
diff=1123.5689-1123.5689
ax1.plot([330.26, 330.3], [330.26+diff, 330.3+diff], ':r')

ax2.plot(df_combo_voigt['pk1_peak_cent'],  De['1122cm-1'], '.k')
ax3.plot(df_combo_voigt['pk2_peak_cent'],  De['1453cm-1'], '.k')
ax2.plot([np.min(De['1122cm-1']), np.max(De['1122cm-1'])],
         [np.min(De['1122cm-1']), np.max(De['1122cm-1'])], '-r')
ax3.plot([np.min(De['1453cm-1']), np.max(De['1453cm-1'])],
         [np.min(De['1453cm-1']), np.max(De['1453cm-1'])], '-r')
ax1.ticklabel_format(useOffset=False)
ax2.ticklabel_format(useOffset=False)
ax3.ticklabel_format(useOffset=False)
ax1.set_xlabel('Splitting python')
ax1.set_ylabel('Splitting fityk')
ax2.set_xlabel('Pk1 python')
ax2.set_ylabel('Pk1 fityk')
ax3.set_xlabel('Pk2 python')
ax3.set_ylabel('Pk2 fityk')
fig.tight_layout()
fig.suptitle('Voigt fit', fontsize=16, x=0.5, y=1.1)
fig.savefig('GL_corr_Voigt.png', dpi=200)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(10,4))

ax1.plot(df_combo_pvoigt['deltaNe'],  De['Ne-m (cm-1)'], '.k')
ax1.plot([330.26, 330.3], [330.26, 330.3], '-r')
diff=1123.5689-1123.5689
ax1.plot([330.26, 330.3], [330.26+diff, 330.3+diff], ':r')

ax2.plot(df_combo_pvoigt['pk1_peak_cent'],  De['1122cm-1'], '.k')
ax3.plot(df_combo_pvoigt['pk2_peak_cent'],  De['1453cm-1'], '.k')
ax2.plot([np.min(De['1122cm-1']), np.max(De['1122cm-1'])],
         [np.min(De['1122cm-1']), np.max(De['1122cm-1'])], '-r')
ax3.plot([np.min(De['1453cm-1']), np.max(De['1453cm-1'])],
         [np.min(De['1453cm-1']), np.max(De['1453cm-1'])], '-r')
ax1.ticklabel_format(useOffset=False)
ax2.ticklabel_format(useOffset=False)
ax3.ticklabel_format(useOffset=False)
ax1.set_xlabel('Splitting python')
ax1.set_ylabel('Splitting fityk')
ax2.set_xlabel('Pk1 python')
ax2.set_ylabel('Pk1 fityk')
ax3.set_xlabel('Pk2 python')
ax3.set_ylabel('Pk2 fityk')
fig.tight_layout()
fig.suptitle('Pseudovoigt fit', fontsize=16, x=0.5, y=1.1)
fig.savefig('GL_corr_Voigt.png', dpi=200)
```
