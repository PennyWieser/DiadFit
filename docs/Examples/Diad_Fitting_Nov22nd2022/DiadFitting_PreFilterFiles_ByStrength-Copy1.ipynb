```python
# This specifies what file type your Raman exported in. 
filetype='headless_txt' # Specifies raman file format 
file_ext='.txt'
prefix=False
```

## This function finds all files in your path with the specified file format
- As you will fit your Ne lines in a different workbook, you can use 'exclude_str' to put any text that appears in all your Ne files, and it won't look here

```python
Diad_files=pf.get_diad_files(path=spectra_path, file_ext=file_ext, 
                            exclude_str='N',
                            exclude_str_2='FG04',
                             exclude_str_3='Sub BG',
                            exclude_str_4='Cap')
Diad_files[0:5]
```

## Filter files by strength

```python
diad_id_config=pf.diad_id_config(height=100,  distance=10, prominence=400)
diad_id_config

# Specify the number of peaks, 1 = just peak, 2 = Peak+hot bands
n_peaks_diad1=2
n_peaks_diad2=3
    
diad1_peaks, diad2_peaks=pf.identify_diad_peaks(
    config=diad_id_config, path=spectra_path, filename=Diad_files[0],
filetype=filetype, n_peaks_diad1=n_peaks_diad1, n_peaks_diad2=n_peaks_diad2)
```

```python
Diad_Files_Grp1,  Diad_Files_Grp2,  Diad_Files_Grp3=pf.filter_files_by_intensity(
Diad_files=Diad_files, spectra_path=spectra_path, filetype=filetype,
combo_upper_cutoff=30000, combo_lower_cutoff=4000, yoff=300)
```

## Now lets save these lists, and we load them in another notebook

```python
import joblib

# Dumps into file
joblib.dump(Diad_Files_Grp1, 'Diad_Files_Grp1.sav')
joblib.dump(Diad_Files_Grp2, 'Diad_Files_Grp2.sav')
joblib.dump(Diad_Files_Grp3, 'Diad_Files_Grp3.sav')
```
